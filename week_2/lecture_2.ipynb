## Building Intuition for Classification

### Understanding Decision Boundaries

**Decision Boundaries:** What are they?
- Decision boundaries are the lines, curves, or surfaces that separate data points into classes. 
- In a 2D feature space, this boundary might be a line. In higher dimensions, it could be a plane or a hyperplane.

Below is a code for visual representation of decision boundaries for some simple classification problem.

```python
#@title Logistic Regression Decision Boundary
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression

# Generate a synthetic binary classification dataset
X, y = make_classification(n_features=2, n_redundant=0, n_informative=2, random_state=1, n_clusters_per_class=1)

# Fit a logistic regression model
model = LogisticRegression()
model.fit(X, y)

# Plotting decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.4)
plt.scatter(X[:, 0], X[:, 1], c=y, s=20, edgecolor='k')
plt.title('Decision Boundary with Logistic Regression')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

**How do we interpret the decision boundary?**
- The decision boundary visually demonstrates how well the model is able to separate the different classes. 
- A clear separation indicates a model that can accurately classify the data points, while a poorly defined boundary might suggest that the model struggles to distinguish between the classes.

Let's see this distinction in action with a simple example using synthetic data and a linear classifier.

```python
#@title Decision Boundary with Clear Separation
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.linear_model import LogisticRegression

# Generate a linearly separable dataset
X, y = make_blobs(n_samples=100, centers=2, n_features=2, cluster_std=1.5, random_state=2)

# Fit the logistic regression model
model_linear = LogisticRegression()
model_linear.fit(X, y)

# Create a mesh to plot the decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500), np.linspace(y_min, y_max, 500))

# Predict classifications over the mesh
Z = model_linear.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

# Plotting
plt.contourf(xx, yy, Z, alpha=0.5)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k')
plt.title('Linearly Separable Data Decision Boundary')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
#@title Decision Boundary with Overlapping Classes
from sklearn.datasets import make_circles

# Generate a circular dataset
X_circular, y_circular = make_circles(n_samples=100, noise=0.05, factor=0.5, random_state=1)

# Fit the logistic regression model
model_circular = LogisticRegression()
model_circular.fit(X_circular, y_circular)

# Create a mesh to plot the decision boundary
x_min, x_max = X_circular[:, 0].min() - 1, X_circular[:, 0].max() + 1
y_min, y_max = X_circular[:, 1].min() - 1, X_circular[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500), np.linspace(y_min, y_max, 500))

# Predict classifications over the mesh
Z = model_circular.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

# Plotting
plt.contourf(xx, yy, Z, alpha=0.5)
plt.scatter(X_circular[:, 0], X_circular[:, 1], c=y_circular, edgecolor='k')
plt.title('Circular Data Decision Boundary')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

**Beyond Linear Boundaries:** What if the data is not linearly separable?
- In many real-world scenarios, data is not linearly separable, meaning a straight line cannot accurately separate the classes.
- In such cases, we need more complex models that can capture non-linear relationships between features and target classes. Some examples include decision trees, support vector machines, and neural networks.

```python
#@title Decision Boundary with Non-linearly Separable Classes with a more complex model
from sklearn.datasets import make_circles
from sklearn.svm import SVC

# Generate a circular dataset
X_circular, y_circular = make_circles(n_samples=100, noise=0.05, factor=0.5, random_state=1)

# Fit the SVM model
model_circular = SVC(kernel='rbf', C=1E6)
model_circular.fit(X_circular, y_circular)

# Create a mesh to plot the decision boundary
x_min, x_max = X_circular[:, 0].min() - 1, X_circular[:, 0].max() + 1
y_min, y_max = X_circular[:, 1].min() - 1, X_circular[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500), np.linspace(y_min, y_max, 500))

# Predict classifications over the mesh
Z = model_circular.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

# Plotting
plt.contourf(xx, yy, Z, alpha=0.5)
plt.scatter(X_circular[:, 0], X_circular[:, 1], c=y_circular, edgecolor='k')
plt.title('Circular Data Decision Boundary with a more complex model')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
```

### Binary vs. Multiclass Classification

**Binary Classification:** What is it?
-  0 or 1, True or False, Yes or No, 2 classes, hence "binary".
- Examples: Spam detection, fraud detection, disease diagnosis, etc.

**Multiclass Classification:** What about more than 2 classes?
- Predicting one class out of multiple classes.
- Examples: Image classification (e.g., cat, dog, bird), sentiment analysis (e.g., positive, negative, neutral), etc.
- Also called multinomial classification.

**Some key considerations:**
- **Algorithm Selection:** Some algorithms are inherently binary classifiers (e.g., logistic regression), while others can handle multiple classes directly (e.g., decision trees, k-NN).
- **One-vs-All (OvA) vs. One-vs-One (OvO):** There are strategies to extend binary classifiers to multiclass problems. 
- **Performance Metrics:** Evaluation metrics like accuracy, precision, recall, F1 score, etc., may need to be adapted for multiclass scenarios.

## Deep Dive to Logistic Regression

### Introduction to Logistic Regression

**What is Logistic Regression?** 
- Logistic regression is a statistical model used for binary classification tasks.
- Can be extended to multiclass problems using techniques like one-vs-rest (OvR).
- Despite its name, logistic regression is used for classification, not regression.
- It models the probability that a given input belongs to a particular class.
- Uses the logistic (sigmoid) function to map input values to probabilities.

**Why Logistic Regression?**
- **Interpretability:** 
    - Logistic regression provides interpretable results, as the coefficients can be directly linked to the impact of each feature on the predicted probability.
    - For example, a positive coefficient indicates that an increase in the feature value leads to a higher probability of the positive class.
- **Probabilistic Output:**
    - Logistic regression outputs probabilities, allowing for a more nuanced understanding of the model's confidence in its predictions.
    - This is particularly useful when decision-making requires a measure of uncertainty.
- **Efficiency:**
    - Logistic regression is computationally efficient and can handle large datasets with ease.
    - It is a good starting point for binary classification tasks before exploring more complex models.

### The Mathematics of Logistic Regression

**From Linear to Logistic Regression:**
- Logistic regression is an extension of linear regression to the classification domain.
- While linear regression predictions can be any real number, logistic regression predictions are probabilities between 0 and 1. 
- The logistic function transforms the linear regression equation so that its output fits within this range.

**Logistic Function and Sigmoid Curve:**
- The logistic function, also known as the sigmoid function, is a key component of logistic regression.
- It maps any real value to the range [0, 1], making it suitable for modeling probabilities.
- Before showing you the equation, let's have a moment to think; can you come up with a function that transforms any real number to a value between 0 and 1?
- Simplest answer could be $f(z) = \frac{1}{1 + n^{z}}$ where $n$ is a positive number. This is very close to the sigmoid function! If we use $n = e$, and put a minus sign on $z$ we get the sigmoid function.

- **Why $e$?** This is not an arbitrary selection, there are some properties of $e$ that makes it a good choice for this function. This would be a quite interesting discussion for a later time!
- Who can explain why $f(z) = \frac{1}{1 + n^{-z}}$ is a good choice for this transformation?
- The sigmoid function is defined as: $S(z) = \frac{1}{1 + e^{-z}}$
    - Where $z$ is the linear combination of input features and model coefficients. $z = w_1x_1 + w_2x_2 + ... + w_nx_n + b$
- The sigmoid function's S-shaped curve reflects how changes in the input affect the predicted probability. 
- For values of $z$ near 0, small changes in $z$ can lead to significant changes in $S(z)$, reflecting uncertainty. 
    - As $z$ moves away from 0, the curve flattens, indicating more confidence in the prediction (Like 99% sure that it is 1 or 0).

```python
#@title Logistic Regression Example with Single Feature
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset: Hours studied vs Pass(1)/Fail(0)
hours_studied = np.array([0.5, 1.5, 2.0, 3.5, 4.0, 4.5, 5.0, 6.0])
pass_fail = np.array([0, 0, 0, 1, 1, 1, 1, 1])

# Plotting the dataset
plt.scatter(hours_studied, pass_fail, color='red', label='Data')

# Assuming a logistic regression model 'model' has been fit to these data
# Let's simulate a fitted model for visualization purposes
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate a range of hours to plot the sigmoid curve
x_values = np.linspace(0, 7, 300)
# Simulated model parameters for visualization
b, w = -6, 2  # Intercept and weight
y_values = sigmoid(w * x_values + b)

plt.plot(x_values, y_values, label='Sigmoid Function')

plt.axhline(y=0.5, color='gray', linestyle='--', label='Decision Threshold')
plt.xlabel('Hours Studied')
plt.ylabel('Probability of Passing')
plt.legend()
plt.show()
```

```python
#@title Logistic Regression Example with Two Features
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification

# Generating a synthetic two-feature dataset
X, y = make_classification(n_features=2, n_redundant=0, n_clusters_per_class=1, n_samples=100, weights=[0.5], flip_y=0, random_state=42)

# Fitting the logistic regression model
model = LogisticRegression()
model.fit(X, y)

# Plotting the dataset
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter', edgecolor='k', label='Data')

# Decision boundary
x_boundary = np.linspace(X[:, 0].min(), X[:, 0].max(), 100)
y_boundary = -(model.coef_[0][0] * x_boundary + model.intercept_) / model.coef_[0][1]

plt.plot(x_boundary, y_boundary, 'r--', label='Decision Boundary')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```


**Modeling Probabilities and Thresholding:**
Logistic regression models the probability that a given input $X$ belongs to the positive class (usually labeled as 1) as $P(Y=1|X) = S(z)$. 
- To make a binary classification decision, a threshold is set (commonly 0.5) - if below 0, if above 1 as the predicted class.
- This threshold can be adjusted depending on the problem's context and the importance of false positives vs. false negatives.

**From Odds to Log-Odds**
In logistic regression, we model the log-odds (logit) of the probability of the positive class:
$$\log\left(\frac{P(Y=1|X)}{1 - P(Y=1|X)}\right) = z = w_1x_1 + w_2x_2 + ... + w_nx_n + b$$

- **Odds:** The odds of an event is the ratio of the probability of the event occurring to the probability of the event not occurring. 
    - In the context of logistic regression, it's how we quantify the **_likelihood_** of a data point belonging to the positive class versus the negative class.
- **Log-Odds (Logit):** Taking the logarithm of the odds simplifies the relationship between the dependent and independent variables, transforming it into a linear relationship that can be modeled with the input features.
    - **_Dependent/Independent Variables:_** In logistic regression, the target variable (dependent) is modeled as a function of the input features (independent).

### Maximum Likelihood Estimation (MLE)

- MLE is a fundamental statistical method used to estimate the parameters of a model. 
- In the context of logistic regression, MLE aims to find the parameter values (coefficients) _that make the observed data most probable_.

**Understanding Likelihood**
- **Likelihood** refers to the probability of observing the given data (the set of observed class labels) under specific model parameters (coefficients and intercept).
- **Intuitively**, think of likelihood as a measure of how well our model explains the observed outcomes. 
    - For a set of parameters, if the predicted probabilities closely match the actual class labels in our training data, the likelihood of those parameters is high.

#### How MLE Works

1. **Modeling Probabilities:** Logistic regression models the probability that each data point belongs to the positive class (1) based on its features. This probability is given by the logistic function:
   
   $$P(Y=1|X;w) = \frac{1}{1 + e^{-(w \cdot X + b)}}$$

2. **Constructing the Likelihood Function:** The likelihood function combines the probabilities for all data points in the dataset. For binary classification, it can be expressed as:
   
   $$L(w) = \prod_{i=1}^{n} P(Y=y_i|X=x_i;w)^{y_i} \cdot (1 - P(Y=y_i|X=x_i;w))^{(1-y_i)}$$
   
   Here, $y_i$ is the actual class label for the $i$ th data point, and $P(Y=y_i|X=x_i;w)$ is the predicted probability that the \(i\)th data point belongs to class 1, based on its features $x_i$ and the model parameters $w$.

3. **Maximizing the Likelihood:** The goal of MLE is to find the parameter values $w$ that maximize the likelihood function $L(w)$. 
    - Since multiplying many probabilities (which are less than 1) can lead to very small numbers, we often maximize the log-likelihood instead, which turns the product into a sum, making the calculation more manageable:
   
   $$\log(L(w)) = \sum_{i=1}^{n} [y_i \cdot \log(P(Y=1|X=x_i;w)) + (1-y_i) \cdot \log(1 - P(Y=1|X=x_i;w))]$$

4. **Optimization:** Finding the parameters that maximize the log-likelihood involves optimization techniques like gradient descent, where we iteratively adjust $w$ to increase the log-likelihood until we reach the maximum.

By maximizing the likelihood, we ensure our model parameters are tuned to best explain the observed data, leading to more accurate predictions. _LEARNING HAPPENS HERE!_

#### Simplified Example

- Imagine you're trying to find the right settings on a machine that sorts apples into "ripe" and "unripe" based on color. 
- You adjust the settings (parameters) and test the machine on a batch of apples whose ripeness you already know. 
- The likelihood is a measure of how many apples are correctly classified by the machine under each setting. 
- MLE involves tweaking these settings until the machine's performance (likelihood of correctly classifying the apples) is maximized.


### Gradient Descent for Optimization

- **What is optimization?** Optimization is the process of finding the best solution (parameters) that minimizes or maximizes a given objective function (e.g., cost function).
- **What is our goal (objective) in logistic regression?** Our goal is to find the parameters (coefficients) that maximize the likelihood of observing the given class labels.
- **How do we achieve this?** We use optimization algorithms like gradient descent to _iteratively_ update the parameters towards the optimal values.

<img src="https://easyai.tech/wp-content/uploads/2019/01/tiduxiajiang-1.png" width="800">

**Optimization and Gradient Descent:**
- Gradient descent is an iterative optimization algorithm used to minimize a function by moving in the direction of the steepest descent (negative gradient).
- It iteratively adjusts the model's parameters (coefficients) to find the set that minimizes the cost, essentially "learning" from the data - just like we mentioned in MLE.
- Super important and useful! - What made Deep Learning so popular.

#### **Understanding the Cost Function in Logistic Regression**

Before diving into gradient descent, it's important to understand the cost function used in logistic regression. 
- The cost function, also known as the log loss or binary cross-entropy loss, quantifies the difference between the predicted probabilities and the actual class labels in the training data. 
- For logistic regression, the cost function is defined as:

$$J(w) = -\frac{1}{m} \sum_{i=1}^{m} [y^{(i)} \log(\hat{y}^{(i)}) + (1 - y^{(i)}) \log(1 - \hat{y}^{(i)})]$$

where:
- $m$ is the number of observations,
- $y^{(i)}$ is the actual class label for observation $i$,
- $\hat{y}^{(i)}$ is the predicted probability that observation $i$ belongs to the positive class, computed as $\hat{y}^{(i)} = S(w^Tx^{(i)} + b)$,
- $w$ represents the model's coefficients, and $b$ is the intercept.

This cost function is chosen because it penalizes incorrect predictions in a way that's well-suited for the probabilistic output of logistic regression. 
- When the actual class is 1, the cost increases sharply as the predicted probability approaches 0. 
- When the actual class is 0, the cost increases sharply as the predicted probability approaches 1.

Check image below!

```python
#@title Cost Function vs. Predicted Probability
import numpy as np
import matplotlib.pyplot as plt

# Generate a range of probabilities from 0 to 1
probabilities = np.linspace(0.01, 0.99, 100)

# Calculate the cost for actual class 1 and 0
cost_actual_1 = -np.log(probabilities)
cost_actual_0 = -np.log(1 - probabilities)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(probabilities, cost_actual_1, label='Actual class = 1')
plt.plot(probabilities, cost_actual_0, label='Actual class = 0', linestyle='--')
plt.title('Cost Function vs. Predicted Probability')
plt.xlabel('Predicted Probability (P(Y=1|X))')
plt.ylabel('Cost')
plt.legend()
plt.grid(True)
plt.show()
```

- Possible Q: Wait, what just happened... Where did the MLE go? How did we end up with this cost function? I thought we were maximizing the likelihood?
    - This is a great question! The cost function in logistic regression is derived from the **negative log-likelihood**, which is equivalent to maximizing the likelihood. 
    - By minimizing the negative log-likelihood, we are effectively maximizing the likelihood of observing the given class labels under the model's predicted probabilities.
    - **Fun little exercise: Can you derive the cost function from the likelihood function we discussed earlier?**

#### **The Gradient Descent Algorithm**

Gradient descent minimizes the cost function by iteratively moving towards the set of parameters that lead to the lowest cost. 

It updates the parameters in the direction opposite to the _gradient of the cost function_ with respect to the parameters:
- $w := w - \alpha \frac{\partial J(w)}{\partial w}$
- $b := b - \alpha \frac{\partial J(b)}{\partial b}$

where:
- $\alpha$ is the learning rate, a hyperparameter that controls the step size during each iteration,
- $\frac{\partial J(w)}{\partial w}$ and $\frac{\partial J(b)}{\partial b}$ are the partial derivatives of the cost function with respect to the coefficients and intercept, respectively, indicating the direction of the steepest ascent.
    - **Partial Derivatives:** The rate of change of a function with respect to one of its variables, treating the other variables as constants.

#### **Implementing Gradient Descent**

- Implementing gradient descent involves calculating these derivatives and using them to update the parameters iteratively.
- The updates continue until the cost function converges to its minimum value, indicating that the model parameters are optimized for the given data.

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Define the cost function and its derivative
def cost_function(x):
    return x ** 2

def derivative(x):
    return 2 * x

# Function to plot the cost function, current point, and gradient step
def plot_gradient_descent(alpha=0.1, start_x=9):
    # Generate x values and corresponding y values for the cost function plot
    x_values = np.linspace(-10, 10, 400)
    y_values = cost_function(x_values)

    # Calculate the next x position after taking a gradient step
    next_x = start_x - alpha * derivative(start_x)

    plt.figure(figsize=(10, 6))
    plt.plot(x_values, y_values, label='Cost Function')
    plt.scatter([start_x], [cost_function(start_x)], color='red', zorder=5, label='Current Position')
    plt.scatter([next_x], [cost_function(next_x)], color='green', zorder=5, label='Next Position After Step')

    # Draw a line representing the gradient step
    plt.plot([start_x, next_x], [cost_function(start_x), cost_function(next_x)], 'k--', zorder=5)

    plt.title('Interactive Gradient Descent Simulation')
    plt.xlabel('w')
    plt.ylabel('f(w)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Create interactive widgets
interact(plot_gradient_descent,
         alpha=FloatSlider(value=0.1, min=0.01, max=1, step=0.01, description='Learning Rate:', readout_format='.2f'),
         start_x=FloatSlider(value=9, min=-10, max=10, step=0.1, description='Start w:', readout_format='.1f'))
```

#### **Challenges and Considerations**

- **Choosing the Learning Rate:** A learning rate that's too high may cause the algorithm to overshoot the minimum, while a learning rate that's too low may result in a long convergence time.

## Feature Handling and Preprocessing (Time Permitting)

### Feature Scaling

- **What is Feature Scaling?** Feature scaling is the process of standardizing the range of independent variables or features of data.

1. **Importance of Feature Scaling in Logistic Regression:**
   - Logistic regression uses gradient descent (or similar optimization algorithms) to find the optimal coefficients that minimize the cost function. 
   - Feature scaling ensures that all features contribute equally to the cost function, preventing features with larger scales from disproportionately influencing the model's predictions.

2. **Methods of Feature Scaling:**
   - **Standardization:** This involves transforming each feature to have a mean of 0 and a standard deviation of 1. 
      - $(x - \mu) / \sigma$ where $\mu$ is the mean and $\sigma$ is the standard deviation of the feature. 
      - Does not bound values to a specific range, which may be preferable for features with outliers or heavy-tailed distributions.
   - **Normalization (Min-Max Scaling):** This approach scales and translates each feature individually such that it is in the range [0, 1]. 
      - The transformation is given by $(x - \text{min}) / (\text{max} - \text{min})$, where $text{min}$ and $\text{max}$ are the minimum and maximum values of the feature, respectively. 
      - Normalization is useful when you want to bound features to a specific range.

```python
#@title Standardization and Min-Max Normalization
import numpy as np
import matplotlib.pyplot as plt

# Generating synthetic data
np.random.seed(0)
data = np.random.randint(0, 100, size=(100, 1))

# Function for Min-Max Normalization
def min_max_scaling(data):
    min_val = np.min(data)
    max_val = np.max(data)
    scaled_data = (data - min_val) / (max_val - min_val)
    return scaled_data

# Function for Standardization
def standardization(data):
    mean = np.mean(data)
    std = np.std(data)
    standardized_data = (data - mean) / std
    return standardized_data

# Applying Min-Max Normalization
normalized_data = min_max_scaling(data)

# Applying Standardization
standardized_data = standardization(data)

# Plotting
plt.figure(figsize=(15, 5))

# Original Data
plt.subplot(1, 3, 1)
plt.hist(data, bins=15, alpha=0.7, label='Original Data')
plt.title('Original Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Normalized Data
plt.subplot(1, 3, 2)
plt.hist(normalized_data, bins=15, alpha=0.7, label='Normalized Data', color='g')
plt.title('Normalized Data Distribution')
plt.xlabel('Normalized Value')
plt.ylabel('Frequency')

# Standardized Data
plt.subplot(1, 3, 3)
plt.hist(standardized_data, bins=15, alpha=0.7, label='Standardized Data', color='r')
plt.title('Standardized Data Distribution')
plt.xlabel('Standardized Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Generating synthetic 1D data
np.random.seed(42)
data_1d = np.random.randint(0, 50, size=100).reshape(-1, 1)

def scale_data(scaler_type='min_max'):
    if scaler_type == 'min_max':
        scaler = MinMaxScaler()
    elif scaler_type == 'standard':
        scaler = StandardScaler()

    data_scaled = scaler.fit_transform(data_1d)

    plt.figure(figsize=(10, 2))
    plt.scatter(data_scaled, np.zeros_like(data_scaled), alpha=0.7, color='green')
    plt.yticks([]) # Hide y-axis ticks
    plt.title(f'{scaler_type.replace("_", " ").capitalize()} Scaler Effect')
    plt.xlabel('Scaled Feature Value')
    plt.grid(True)
    plt.xlim(-2, 2) # StandardScaler might have values beyond [0, 1]
    plt.show()

# Setting up the interactive widget
interact(scale_data,
         scaler_type=['min_max', 'standard'])
```


#### Handling Categorical Variables

**Why Categorical Variables Need Special Treatment:**
   - Logistic regression and many other machine learning algorithms require numerical input. 
   - Categorical variables, which represent discrete values or categories, must be converted into a numerical format that preserves the information without introducing ordinal relationships (unless such relationships exist).

**Techniques for Encoding Categorical Variables:**
   - **One-Hot Encoding:** This method converts each category value into a new binary variable (0 or 1). 
       - For each observation, only one of these binaries will be "hot" (1), indicating the presence of a particular category, while the rest will be "cold" (0). 
       - One-hot encoding is beneficial because it does not introduce an artificial ordering of categories. _However, it can significantly increase the dimensionality of the dataset._
       - An example: If we have a categorical variable "Color" with values "Red," "Green," and "Blue,".
   - **Ordinal Encoding:** If the categorical variable possesses a natural order (ordinal), then ordinal encoding can be applied.
       - Here, each unique category value is assigned an integer value. 
       - This method should be used cautiously, as it implies an ordered relationship among the categories, which may not be appropriate for all features.
       - An example: If we have a categorical variable "Size" with values "Small," "Medium," and "Large,".

**When to Use these guys?**
- In practice, preprocessing steps like feature scaling and encoding categorical variables are implemented as part of the _data preparation_ process before training the model.
Sklearn for easy implementation:
    - `StandardScaler` for standardization
    - `MinMaxScaler` for normalization
    - `OneHotEncoder` for one-hot encoding (many more options available!)

```python
colors = ["Red", "Green", "Blue", "Green", "Red", "Blue"]

# Unique categories
unique_colors = list(set(colors))

# One-hot encode
one_hot_encoded = []
for color in colors:
    encoded = [int(color == unique) for unique in unique_colors]
    one_hot_encoded.append(encoded)

# Visualize
print(f"One-Hot Encoded Data (Columns: {unique_colors}):")
for item in one_hot_encoded:
    print(item)
```

### Model Evaluation Metrics

- We trained the model, now what? How do I know how good it is?
- After training a logistic regression model (or any classification model), it's crucial to evaluate its performance accurately. 
- Evaluation metrics provide insight into how well the model makes predictions, but choosing the right metric is key.

We start with the most straightforward metric, accuracy.

#### Starting with Accuracy

Accuracy is the simplest and most intuitive metric for evaluating classification models. 

It calculates the proportion of correctly predicted observations to the total observations:

$\text{Accuracy} = \frac{\text{Number of Correct Predictions}}{\text{Total Number of Predictions}}$

While accuracy can provide a quick overview of model performance, it may not always be the best metric, especially in cases where the data is imbalanced (i.e., when the number of observations in different classes significantly varies).

**Q: Why do you think accuracy might not be the best metric for imbalanced datasets?**

#### Beyond Accuracy

While accuracy might seem like an ideal metric due to its simplicity, it doesn't tell the whole story, 
especially in imbalanced datasets or when the cost of false positives differs significantly from false negatives. 
Here, we explore additional metrics that offer a more comprehensive evaluation.

##### Precision and Recall

- **Precision** (Positive Predictive Value) measures the proportion of true positive predictions in the positive class's predictive values. It answers the question, "Of all the instances the model labeled as positive, how many are actually positive?"

$\text{Precision} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Positives}}$

- **Recall** (Sensitivity or True Positive Rate) measures the proportion of actual positives that were correctly identified by the model. It addresses the question, "Of all the positive instances, how many did the model correctly identify?"

$\text{Recall} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Negatives}}$

- **F1 Score** is the harmonic mean of precision and recall, offering a balance between the two by penalizing extreme values. It's particularly useful when you need to balance precision and recall.

$\text{F1 Score} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}$

- What is "Positive", "Negative", "True", "False" in this context? 
    - **Positive:** The class we are interested in predicting (e.g., disease, fraud, spam).
    - **Negative:** The other class (e.g., not disease, not fraud, not spam).
    - **True:** Correctly predicted observations.
    - **False:** Incorrectly predicted observations.

**Q: So if we are predicting a disease, what would be a positive prediction? What about a true positive prediction?**

##### Confusion Matrix

The confusion matrix is a table used to describe the performance of a classification model on a set of test data for which the true values are known. It provides insight into the types of errors made by the model.

- **True Positives (TP):** Correctly predicted positive observations
- **True Negatives (TN):** Correctly predicted negative observations
- **False Positives (FP):** Incorrectly predicted positive observations (Type I error)
- **False Negatives (FN):** Incorrectly predicted negative observations (Type II error)

##### ROC Curve and AUC

- **ROC Curve (Receiver Operating Characteristic Curve):** Plots the true positive rate (recall) against the false positive rate (1 - specificity) at various threshold settings. The curve illustrates the trade-off between sensitivity and specificity.
    - **Sensitivity:** The true positive rate, i.e., the proportion of actual positives that are correctly identified.
    - **Specificity:** The true negative rate, i.e., the proportion of actual negatives that are correctly identified.
    - **Threshold:** The value above which the model predicts the positive class.
- **AUC (Area Under the ROC Curve):** Summarizes the ROC curve into a single value, representing the model's ability to discriminate between the positive and negative classes. An AUC of 1 represents a perfect model, while an AUC of 0.5 represents a no-skill classifier.
